<h3>Simple Linear Regression </h3>This is the foundation of many regression algorithms. It models the relationship between a dependent variable (what you want to predict) and one or more independent variables (features used for prediction) using a linear equation, y=mx +c .Where m is the slope or  gradient of a graph of y/x and c is the intecept on the y-axis. Both m and c are constant for a line plotted from pool of data points.



From the equation, the value of y can be predicted for any given value of x provided the constant values m and c are known.
Let us use numpy, a python library to create array of data and matplotlib which is another library to plot graph to illustrate.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,10,15,20,25,30]) 
y = np.array([13,23,33,43,53,63]) 

plt.plot(x, y)
plt.show()

```

Instead of manual plotting,the slope of the line that best fits of the data can be obtained using linear regression from the scipy.stats library:

```python
from scipy.stats import linregress
slope, intercept, r_value, p_value, std_err = linregress(x, y)
print("Slope m:", slope)
print("Intercept c:", intercept)
std_err
```

Since values of m and c are known, we can predict , from  equation y= mx+c, the value of  y when any x is given  as:<br>
